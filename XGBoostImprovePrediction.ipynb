## Import data from PDFs

```python
import requests
import pdfplumber

# URLs of the PDF files
pdf_urls = [
    'https://www.vintercupen.nu/wp-content/uploads/2023/01/VC-2023-1_6-Heat-1-Totalt.pdf',
    'https://www.vintercupen.nu/wp-content/uploads/2023/01/VC-2023-1_6-Heat-2-Totalt.pdf',
    'https://www.vintercupen.nu/wp-content/uploads/2023/01/VC-2023-1_6-Heat-3-Totalt.pdf'
]

# Aggregate text from all PDFs
pdf_text = ""

for url in pdf_urls:
    # Fetch the PDF file from the URL
    response = requests.get(url)
    
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Save the content of the response to a file
        with open('downloaded_pdf.pdf', 'wb') as f:
            f.write(response.content)

        # Open the downloaded PDF file
        with pdfplumber.open('downloaded_pdf.pdf') as pdf:
            # Extract text from each page
            for page in pdf.pages:
                pdf_text += page.extract_text()
    else:
        print(f"Failed to fetch the PDF file from the URL: {url}")

# Display the aggregated text from all PDFs
print(pdf_text)
```

```python
lines = pdf_text.split('\n')
for i, line in enumerate(lines):
    print(f"Line {i + 1}: {line}")
```

## Make text to dataframe.

```python
import pandas as pd
from io import StringIO

# Replace spaces with commas to create a CSV-like string
csv_like_text = pdf_text.replace(' ', ',')

# Create a DataFrame
df = pd.read_csv(StringIO(csv_like_text), sep=',', skipinitialspace=True)

# Display the DataFrame
print(df)
```
