```python
import os
import re
import pandas as pd
import numpy as np
```

```python
# Folder containing CSV files
folder_path = 'lap_times_fixed/'

# List all files in the folder ending with '.csv'
file_paths = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

# Initialize a dictionary to store dataframes
dataframes = {}

# Loop through each file and read its data into a pandas DataFrame
for file_path in file_paths:
    # Extract file name without extension
    file_name = os.path.splitext(os.path.basename(file_path))[0]
    
    # Read CSV into a dataframe and store it in the dictionary
    dataframes[file_name] = pd.read_csv(file_path)
```

Convert lap times to time format.

```python
# Loop through each dataframe
for key in dataframes.keys():
    df = dataframes[key]
    
    # Find only columns starting with 'Varv' followed by a space and a number
    varv_columns = [col for col in df.columns if re.match(r'Varv\s\d+', col)]
    
    # Replace 'BRUTIT' with NaN in the 'Varv' columns
    df[varv_columns] = df[varv_columns].replace('BRUTIT', pd.NaT)
    
    # Function to convert lap time strings to Timedelta objects
    def convert_to_timedelta(lap_time):
        if pd.notnull(lap_time) and isinstance(lap_time, str) and ':' in lap_time:
            minutes, rest = lap_time.split(':')
            seconds_tenths = rest.split('.')
            if len(seconds_tenths) == 2:
                seconds, tenths = map(int, seconds_tenths)
                microseconds = tenths * 100000  # Convert tenths to microseconds
                return pd.Timedelta(minutes=int(minutes), seconds=seconds, microseconds=microseconds)
        return pd.NaT if pd.isnull(lap_time) else lap_time
    
    # Apply the conversion function to lap time columns
    for col in varv_columns:
        df[col] = df[col].apply(convert_to_timedelta)
    
    # Display the modified DataFrame with 'Varv' columns converted to Timedelta
print(df.head(3))
```

## Drop slowest time for each row and calculate average of the rest.

It collects lap times from the specified columns (varv_columns) in each row.
It filters out non-Timedelta values from these lap times.
It excludes the slowest time (highest value) from the valid lap times.
It calculates the average of the remaining lap time
s.
Add new column with average lap time.

```python
# Loop through each dataframe
for key in dataframes.keys():
    df = dataframes[key]
    
    # Calculate average lap times excluding the slowest time
    def calculate_average(row):
        lap_times = row[varv_columns].dropna().values
        valid_lap_times = [time for time in lap_times if isinstance(time, pd.Timedelta)]
        if len(valid_lap_times) >= 2:
            slowest = max(valid_lap_times)
            lap_times = [time for time in valid_lap_times if time != slowest]
            return sum(lap_times, pd.Timedelta(0)) / len(lap_times)
        return pd.NaT
    
    # Create a new column 'Snitt Varvtid' with the average lap times
    df['Snitt Varvtid'] = df.apply(calculate_average, axis=1)
    
    # Display the modified DataFrame
print(df.head(-3))
```

Sort the list ascending upon "Snitt Varvtid" into new dataframes.

```python
sorted_dataframes = {}

for key, df in dataframes.items():
    sorted_dataframes[key] = df.sort_values(by='Snitt Varvtid', ascending=True)
```

```python
# Get the unique 'Namn' values
unique_names = set()
for df in sorted_dataframes.values():
    unique_names.update(df['Namn'])

# Create the ranked_dataframe
ranked_dataframe = pd.DataFrame({'Namn': list(unique_names)})

# Merge based on 'Namn' column
for idx, (key, df) in enumerate(sorted_dataframes.items(), start=1):
    col_name = df.columns[0]
    df_subset = df[['Namn', 'Rank', 'Klass', 'Klubb']]
    df_subset = df_subset.rename(columns={'Rank': f'Rank {col_name}', 'Klass': f'Klass_{idx}', 'Klubb': f'Klubb_{idx}'})
    ranked_dataframe = pd.merge(ranked_dataframe, df_subset, on='Namn', how='left')

# Function to merge 'Klass' columns while retaining unique values
def merge_klass(row):
    klasses = [row[f'Klass_{idx}'] for idx in range(1, len(sorted_dataframes) + 1)]
    unique_klasses = set(k for k in klasses if pd.notnull(k))
    return ', '.join(unique_klasses)

# Function to merge 'Klubb' columns while retaining unique values
def merge_klubb(row):
    klubbs = [row[f'Klubb_{idx}'] for idx in range(1, len(sorted_dataframes) + 1)]
    unique_klubbs = set(k for k in klubbs if pd.notnull(k))
    return ', '.join(unique_klubbs)
    

# Apply the function to create the 'Klass' and 'Klubb' columns in the ranked_dataframe
ranked_dataframe['Klass'] = ranked_dataframe.apply(merge_klass, axis=1)
ranked_dataframe['Klubb'] = ranked_dataframe.apply(merge_klubb, axis=1)

# Drop the individual 'Klass' and 'Klubb' columns for each dataset
for idx in range(1, len(sorted_dataframes) + 1):
    ranked_dataframe.drop(f'Klass_{idx}', axis=1, inplace=True)
    ranked_dataframe.drop(f'Klubb_{idx}', axis=1, inplace=True)

ranked_dataframe.head(10)
```

```python
# Initialize a dictionary to hold data for each unique combination of Namn, #, and Klubb
unique_entries = {}

# Iterate through sorted_dataframes
for idx, (key, df) in enumerate(sorted_dataframes.items(), start=1):
    col_name = df.columns[0]
    # Loop through each row in the dataframe
    for index, row in df.iterrows():
        # Create a unique identifier for the combination of Namn, #, and Klubb
        key = (row['Namn'], row['#'], row['Klubb'])
        
        # If the key already exists, update the ranking values for the current dataframe
        if key in unique_entries:
            for col in df.columns:
                if col.startswith('Rank'):
                    unique_entries[key][f'Rank {col_name}'] = min(unique_entries[key].get(f'Rank {col_name}', float('inf')), row[col])
        else:
            # If the key does not exist, create a new entry
            new_entry = {'Namn': row['Namn'], 'Klubb': row['Klubb'], '#': row['#'], 'Klass': row['Klass']}
            for col in df.columns:
                if col.startswith('Rank'):
                    new_entry[f'Rank {col_name}'] = row[col]
            unique_entries[key] = new_entry

# Create a dataframe from the unique_entries dictionary
ranked_dataframe = pd.DataFrame(list(unique_entries.values()))

# Sort the dataframe by 'Namn' or any other desired column
ranked_dataframe.sort_values(by='Namn', inplace=True)

# Display the resulting ranked_dataframe
ranked_dataframe.head(-10)
```

Set average rank.

```python
# Get columns related to Rank in the ranked_dataframe
rank_columns = [col for col in ranked_dataframe.columns if col.startswith('Rank')]

# Create a new column 'AvgRank' with the average of all Rank columns excluding the highest value
def calculate_average_rank(row):
    ranks = row[rank_columns].dropna()
    if len(ranks) >= 2:  # Check if there are at least 2 non-NaN values
        max_rank = np.max(ranks)  # Find the highest rank value
        return np.mean(ranks[ranks != max_rank])  # Calculate the mean excluding the highest rank
    elif len(ranks) == 1:  # If only one non-NaN value
        return ranks.values[0]  # Return the single value as the average rank
    else:
        return np.nan  # Return NaN if no valid ranks

ranked_dataframe['AvgRank'] = ranked_dataframe.apply(calculate_average_rank, axis=1)

ranked_dataframe.head(10)
```

Sort by AvgRank column.

```python
ranked_dataframe = ranked_dataframe.sort_values(by=['AvgRank', '#'])
ranked_dataframe.head(20)
```

## Drop duplicate rows.
Find duplicate rows based on 'Namn' and 'Klubb' columns.
If no duplicate rows is found show message. Else show what rows where dropped.

```python
duplicate_rows = ranked_dataframe[ranked_dataframe.duplicated(subset=['Namn', 'Klubb'], keep='first')]

# Drop duplicates based on 'Namn' and 'Klubb' columns, keeping the first occurrence (lowest 'AvgRank')
ranked_dataframe.drop_duplicates(subset=['Namn', 'Klubb'], keep='first', inplace=True)

# Display the duplicate rows that were dropped
if duplicate_rows.empty:
    print("No duplicate Rows found")
else:
    print("Duplicate Rows being dropped:")
    print(duplicate_rows)
```

Save the ranked dataframe to a csv.

```python
# Create a directory if it doesn't exist
folder_name = 'ranked'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Define the file path for the CSV
file_path = os.path.join(folder_name, 'ranked.csv')

# Save the DataFrame to a CSV file
try:
    ranked_dataframe.to_csv(file_path, index=False)
    print(f"Ranked data saved successfully to '{file_path}'")
except Exception as e:
    print(f"Error saving ranked data: {e}")
```
